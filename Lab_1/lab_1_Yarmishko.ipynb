## Імпорт та візуалізація зображення

```python
import cv2
import matplotlib.pyplot as plt

image_bgr = cv2.imread("saul.jpg")

# Перевірка, чи зображення завантажилось
if image_bgr is None:
    print("Помилка: зображення не знайдено")
else:
    print("Розмір зображення:", image_bgr.shape)

    # переводимо в RGB
    image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(4, 6))
    plt.imshow(image_rgb)
    plt.title("Кольорове зображення")
    plt.axis("off")
    plt.show()
```

```python
# Перетворення у напівтонове зображення
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(4, 6))
plt.imshow(image_gray, cmap="gray")
plt.title("Напівтонове зображення")
plt.axis("off")
plt.show()
```

Зображення імпортовано засобами бібліотеки OpenCV, після чого виконано
перетворення з кольорового простору BGR у RGB для коректної візуалізації
та у напівтоновий формат (grayscale).

```python
# Запис напівтонового зображення на диск
output_filename = "saul_gray.png"

saved = cv2.imwrite(output_filename, image_gray)

print("Зображення збережено:", saved)
print("Файл:", output_filename)
```

## Контрольні запитання

**1. Що таке pip і conda та чим вони відрізняються між собою?**

`pip` — це стандартний інструмент для встановлення бібліотек Python. Він дозволяє
встановлювати пакети, але не керує віртуальними середовищами.

`conda` — це інструмент, який використовується для створення та управління
віртуальними середовищами, а також для встановлення бібліотек.

Основна відмінність полягає в тому, що `pip` працює тільки з пакетами,
а `conda` — і з пакетами, і з віртуальними середовищами.

**2. Як буде виглядати виконання пунктів 1–3 індивідуального завдання за допомогою Virtualenv?**

При використанні `Virtualenv` створюється та активується віртуальне середовище.
Після цього всі необхідні бібліотеки встановлюються за допомогою `pip`.

На відміну від `conda`, `Virtualenv` не має власного менеджера пакетів і не дозволяє
керувати версіями Python.

**3. Як виконати візуалізацію зображення засобами matplotlib?**

Для візуалізації зображення використовується бібліотека `matplotlib`.
Зображення відображається за допомогою функції `plt.imshow()`.

Для напівтонового зображення додатково вказується параметр `cmap="gray"`.
